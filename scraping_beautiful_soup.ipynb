# BeautifulSoup

Scraping of Wikipedia. This is the start of building an intial dataset for a larger analysis of parking and driving through the United States.

```python
from __future__ import print_function, division
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Traffic_congestion'
response = requests.get(url)
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
# soup.find('ol')
```

```python
city_list = soup.find("ol")
```

```python
city_list.find_all('li')[0].text
```

```python
new_list = []

i =0
while i < 65:
    city = city_list.find_all('li')[i].text
    new_list.append(city)
    i = i + 1
    
```

```python
new_list
```

```python
df = pd.DataFrame(new_list)
```

```python
df['city'], df['state'] = df[0].str.split(',', 1).str
```

```python
df['state_1'], df['hours'] = df['state'].str.split(':', 1).str
```

```python
df.head()
```

```python
df = df.drop([0, 'state'], axis=1)
```

```python
df = df.rename({'stae_1':'state'})
```

```python
import pickle

with open('df_beautiful_soup_demo.pkl', 'wb') as f:
    pickle.dump(df, f)
```
