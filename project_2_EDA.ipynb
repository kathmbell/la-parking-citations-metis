```python
from uszipcode import Zipcode
# from __future__ import print_function
from uszipcode import SearchEngine, SimpleZipcode, Zipcode
```

```python
zipcode_list = []
population_list = []
population_density_list = []
housing_units_list = []
occupied_housing_list = []
median_home_value_list = []
median_household_income_list = []

search = SearchEngine(simple_zipcode=True)

def find_zip(x):
        result = search.by_coordinates(x.lat_google, x.long_google, radius=10, returns=5)
        first_result = result[0]
#         pp.pprint(first_result)
    # - zipcode
        zipcode = first_result.zipcode
        zipcode_list.append(zipcode)
    # - population
        population = first_result.population
        population_list.append(population)
    # - population_density
        population_density = first_result.population_density
        population_density_list.append(population_density)
    # - housing units
        housing_units = first_result.housing_units
        housing_units_list.append(housing_units)
    # - occupied_housing_units
        occupied_housing_units = first_result.occupied_housing_units
        occupied_housing_list.append(occupied_housing_units)
    # - median home value
        median_home_value = first_result.median_home_value
        median_home_value_list.append(first_result.median_home_value)
    # - median household income
        median_household_income = first_result.median_household_income
        median_household_income_list.append(median_household_income)
    
df_update.apply(find_zip, axis=1)
# zipcode_list, population_list, population_density_list, housing_units_list, occupied_housing_list, median_home_value_list, median_household_income_list
```
