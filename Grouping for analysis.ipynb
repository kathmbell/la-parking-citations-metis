```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('complete_data_join.pkl', 'rb') as f:
    df = pickle.load(f)
```

```python
df.info()
```

```python
df = df.drop(['issue_date', 'meter_id', 'marked_time', 'marked_time','rp_state_plate'
         ,'plate_expiry_date','vin','body_style','location','route','agency','latitude','longitude'
         ,'population_density','city_name','emp_flag','employment_noiseflag','q1_noiseflag'
         ,'annual_payroll_noiseflag','city','state'], axis=1)

```

```python
df.shape
```

```python
df.info()
```

```python
# - groupby zipcode, day of week, count ticket_number

df_group = df.groupby(['zipcode','weekday_name','population','housing_units'
                       ,'occupied_housing','median_home_value','median_household_income','cbp_employees',
                       'q1_payroll_thousands','annual_payroll_thousands','cbp_number_establishments'])['ticket_number'].count().reset_index()

```

```python
df_group
```

```python
df_group.rename(columns={"ticket_number": "ticket_count"}, inplace=True)
```

```python
df_group.head()
```

```python
df_group_dummy = df_group
```

```python
dummy_zip = pd.get_dummies(df_group_dummy['zipcode'])
```

```python
dummy_zip.head()
```

```python
dummy_weekday = pd.get_dummies(df_group_dummy['weekday_name'])
```

```python
dummy_weekday.head()
```

```python
df_group_1 = df_group
```

```python
df_group_1_week = df_group_1.merge(dummy_weekday, left_index=True,right_index=True)
```

```python
df_group_1_week_zip = df_group_1_week.merge(dummy_zip, left_index=True,right_index=True)
```

```python
df_group_1_week_zip.shape
```

```python
with open('data_with_dummies.pkl', 'wb') as f:
    pickle.dump(df_group_1_week_zip, f)
    
```

```python
with open('data_no_dummies.pkl', 'wb') as f:
    pickle.dump(df_group, f)
```
